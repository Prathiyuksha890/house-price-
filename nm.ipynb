```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler

# Load the dataset
file_path = '/mnt/data/house-prices-100.xlsx'
df = pd.read_excel('/content/house-prices-100.xlsx')

df.drop_duplicates(inplace=True)

# Handle outliers in 'SqFt' using IQR method
Q1 = df['SqFt'].quantile(0.25)
Q3 = df['SqFt'].quantile(0.75)
IQR = Q3 - Q1
lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR
df['SqFt'] = df['SqFt'].clip(lower, upper)

df['Luxury_Index'] = (df['Bathrooms'] + df['Bedrooms']) * df['Brick'].map({'Yes': 1, 'No': 0})
df['Price_per_sqft'] = df['Price'] / df['SqFt'].replace(0, np.nan)

scaler = MinMaxScaler()
num_cols = ['SqFt', 'Bedrooms', 'Bathrooms', 'Price', 'Price_per_sqft']
df[num_cols] = scaler.fit_transform(df[num_cols])

print(df.info())
print(df.describe())
df.drop_duplicates(inplace=True)

# Visualization Sectionplt
plt.figure(figsize=(12, 8))
sns.histplot(df['Price'], bins=30, kde=True, color='teal')
plt.title('Distribution of House Prices')
plt.show()

plt.figure(figsize=(12, 8))
sns.heatmap(df[num_cols].corr(), annot=True, cmap='Blues', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()

sns.pairplot(df[num_cols])
plt.suptitle('Pairplot of Key Features', y=1.02)
plt.show()

plt.figure(figsize=(12, 8))
sns.boxplot(x='Bedrooms', y='Price', data=df, palette='viridis')
plt.title('Price Distribution Across Bedroom Count')
plt.show()

plt.figure(figsize=(12, 8))
sns.scatterplot(x='SqFt', y='Price', hue='Brick', data=df, palette='viridis')
plt.title('Price vs SqFt by Brick Status')
plt.show()

```
